## Cincia de dados
</div>

Com o Colab, voc pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A clula de cdigos abaixo usa <strong>numpy</strong> para gerar dados aleatrios e <strong>matplotlib</strong> para visualiz-los. Para editar o cdigo, basta clicar na clula e comear a editar.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
