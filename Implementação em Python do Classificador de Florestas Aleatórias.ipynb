```
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

# Gerar um conjunto de dados de exemplo
X, y = make_classification(n_samples=1000, n_features=4, n_informative=2, n_redundant=0, random_state=0, shuffle=False)

# Criar e treinar um modelo RandomForestClassifier
clf = RandomForestClassifier(n_estimators=5, max_depth=2, random_state=0)
clf.fit(X, y)

# Fazer uma previsão com o modelo treinado
prediction = clf.predict([[0, 0, 0, 0]])
print(f"Previsão para a entrada [0, 0, 0, 0]: {prediction}")

# Visualizar as árvores de decisão individuais
for i, tree_in_forest in enumerate(clf.estimators_):
    plt.figure(figsize=(20,10))
    plot_tree(tree_in_forest, filled=True, feature_names=[f'Feature {i}' for i in range(X.shape[1])], class_names=['Class 0', 'Class 1'])
    plt.title(f'Árvore de decisão {i+1} no RandomForest')
    plt.show()
```

```
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

# Gerar um conjunto de dados de exemplo
X, y = make_classification(n_samples=1000, n_features=4, n_informative=2, n_redundant=0, random_state=0, shuffle=False)

# Criar e treinar um modelo RandomForestClassifier
clf = RandomForestClassifier(n_estimators=5, max_depth=2, random_state=0)
clf.fit(X, y)

# Fazer uma previsão com o modelo treinado
prediction = clf.predict([[0, 0, 0, 0]])
print(f"Previsão para a entrada [0, 0, 0, 0]: {prediction}")

# Visualizar as árvores de decisão individuais com tamanho reduzido
for i, tree_in_forest in enumerate(clf.estimators_):
    plt.figure(figsize=(10, 5))  # Ajustar o tamanho da figura para ser mais compacto
    plot_tree(tree_in_forest, filled=True, feature_names=[f'Feature {j}' for j in range(X.shape[1])], class_names=['Class 0', 'Class 1'])
    plt.title(f'Árvore de decisão {i+1} no RandomForest')
    plt.show()
```

## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
